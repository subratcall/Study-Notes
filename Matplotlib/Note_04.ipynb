**Errorbars**

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
plt.style.use('seaborn-whitegrid')
```

```python
x = np.linspace(0,10,30)
y = np.cos(x) + 0.6 * np.random.rand(30)

plt.errorbar(x,y,yerr=0.6,fmt='o',color='k',ecolor='lightgray',elinewidth=3,capsize=0)
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import DotProduct,WhiteKernel

x = np.array([1,3,5,7,9])
y = x * np.sin(x)
kernel = DotProduct() + WhiteKernel()
gpr = GaussianProcessRegressor(kernel=kernel,random_state=0)
gpr.fit(x[:,np.newaxis],y)

x_test = np.linspace(0,10,1000)
y_pred,std = gpr.predict(x_test[:,np.newaxis],return_std=True)
```

```python
plt.plot(x,y,'or')
plt.plot(x_test,y_pred,'-',color='gray')

plt.fill_between(x_test,y_pred-std,y_pred+std,color='gray',alpha=0.2)
```
