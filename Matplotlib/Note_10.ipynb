**Seaborn**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
plt.style.use('classic')
```

```python
rng = np.random.RandomState(123)
x = np.linspace(0,10,500)
y = np.cumsum(rng.randn(500,6),0)

plt.plot(x,y)
plt.legend('ABCDEF',ncol=2,loc='upper left')
```

```python
sns.set()

plt.plot(x,y)
plt.legend('ABCDEF',ncol=2,loc='upper left')
```

```python
data = rng.multivariate_normal([0,0],[[5,2],[2,2]],size=2000)
data = pd.DataFrame(data,columns=['x','y'])

plt.hist(data['x'],density=True,alpha=0.5)
plt.hist(data['y'],density=True,alpha=0.5)
```

```python
sns.kdeplot(data['x'],shade=True)
sns.kdeplot(data['y'],shade=True)
```

```python
sns.distplot(data['x'])
sns.distplot(data['y'])
```

```python
sns.kdeplot(data['x'],data['y'],cmap='viridis')
```

```python
with sns.axes_style('white'):
    sns.jointplot('x','y',data,kind='kde',cmap='RdBu')
```

```python
with sns.axes_style('white'):
    sns.jointplot('x','y',data,kind='hex',cmap='viridis')
```

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris,hue='species',height=2.5)
```

```python
tips = sns.load_dataset('tips')
tips['tip_pct'] = 100*tips['tip']/tips['total_bill']
grid = sns.FacetGrid(tips,row='sex',col='time',margin_titles=True)
grid.map(plt.hist,'tip_pct',bins=np.linspace(0,40,15))
```

```python
with sns.axes_style(style='ticks'):
    g = sns.catplot("day", "total_bill", "sex", data=tips,kind='box')
    g.set_axis_labels("Day", "Total Bill");
```

```python
sns.jointplot("total_bill", "tip", data=tips, kind='reg')
```

```python
sns.violinplot("sex", "tip", data=tips,
               palette=["lightblue", "lightpink"])
```

```python
g = sns.lmplot("total_bill", "tip", col='sex', data=tips,
               markers=".", scatter_kws=dict(color='c'))
g.map(plt.axhline, y=0.1, color="k", ls=":")
```
