**Subplots**

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use('seaborn-white')
```

```python
ax1 = plt.axes()
ax2 = plt.axes([0.65,0.65,0.2,0.2])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.5,0.8,0.4],xticklabels=[],ylim=(-1.2,1.2))
ax2 = fig.add_axes([0.1,0.1,0.8,0.4],ylim=(-1.2,1.2))

x = np.linspace(0,10)
ax1.plot(x,np.sin(x),'-g')
ax2.plot(x,np.sin(x),'-r')
```

```python
fig = plt.figure()
for i in range(1,7):
    ax = fig.add_subplot(2,3,i)
    ax.text(0.5,0.5,str((2,3,i)),fontsize=18,ha='center')
fig.subplots_adjust(hspace=0.4,wspace=0.4)
```

```python
fig,ax = plt.subplots(2,3,sharex='col',sharey='row')
```

```python
mean = [0,0]
cov = [[1,1],[1,2]]
rng = np.random.RandomState(123)
x,y = rng.multivariate_normal(mean,cov,3000).T

fig = plt.figure(figsize=(6,6))
grid = plt.GridSpec(4,4,hspace=0.2,wspace=0.2)
ax = fig.add_subplot(grid[:-1,1:])
x_hist = fig.add_subplot(grid[-1,1:],yticklabels=[],sharex=ax) #row and column
y_hist = fig.add_subplot(grid[:-1,0],xticklabels=[],sharey=ax)

ax.plot(x,y,'oc',markersize=4,alpha=0.3)
x_hist.hist(x,40,histtype='stepfilled',orientation='vertical',color='r')
x_hist.invert_yaxis()
y_hist.hist(y,40,histtype='stepfilled',orientation='horizontal',color='g')
y_hist.invert_xaxis()
```
