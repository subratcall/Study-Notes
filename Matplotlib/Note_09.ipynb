**3D Plotting**

```python
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```python
ax = plt.axes(projection='3d')

zline = np.linspace(0,15,1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline,yline,zline,'gray')

rng = np.random.RandomState(123)
zdata = 10*rng.randn(100)
xdata = np.sin(zdata) + 0.1*rng.randn(100)
ydata = np.cos(zdata) + 0.1*rng.randn(100)
ax.scatter3D(xdata,ydata,zdata,c=zdata,cmap='viridis')
```

```python
def f(x,y):
    return np.sin(np.sqrt(x**2+y**2))

x = np.linspace(-6,6,30)
y = np.linspace(-6,6,30)

X,Y = np.meshgrid(x,y)
Z = f(X,Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X,Y,Z,50,cmap='plasma')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

```python
ax.view_init(60,35) #  60 degrees above the x-y plane, rotated 35 degrees counter-clockwise about the z-axis
fig
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_wireframe(X,Y,Z,cmap='RdBu')
ax.set_title('wireframe')
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,
               cmap='viridis')
ax.set_title('surface')
```

```python
theta = 2 * np.pi * np.random.random(1000)
r = 6 * np.random.random(1000)
x = np.ravel(r * np.sin(theta))
y = np.ravel(r * np.cos(theta))
z = f(x, y)

ax = plt.axes(projection='3d')
ax.scatter3D(x,y,z,c=z,cmap='hot',linewidth=0.5)
```

```python
ax = plt.axes(projection='3d')
ax.plot_trisurf(x,y,z,cmap='hot')
```
