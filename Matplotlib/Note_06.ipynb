**Histograms**

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
plt.style.use('fivethirtyeight')
```

```python
rng = np.random.RandomState(123)
data = rng.rand(1000)
plt.hist(data)
```

```python
plt.hist(data,bins=30,density=True,alpha=0.5,histtype='stepfilled',
         color='steelblue',edgecolor='k')
```

```python
x1 = rng.normal(0,0.8,1000)
x2 = rng.normal(-2,1,1000)
x3 = rng.normal(3,2,1000)
kwargs = dict(bins=30,density=True,alpha=0.5,histtype='stepfilled')

plt.hist(x1,**kwargs)
plt.hist(x2,**kwargs)
plt.hist(x3,**kwargs)
```

```python
counts,bin_edges = np.histogram(data,bins=5)
print(counts)
print(bin_edges)
```

________

```python
mean = [0,0]
cov = [[1,1],[1,2]]
x, y = rng.multivariate_normal(mean,cov,10000).T

plt.hist2d(x,y,bins=30,cmap=plt.cm.RdBu,interpolation='nearest')
colorbar = plt.colorbar()
colorbar.set_label('counts in bin')
```

```python
counts,xedges,yedges = np.histogram2d(x,y,bins=5)
print(counts)
print(xedges)
print(yedges)
```

```python
plt.hexbin(x,y,gridsize=30,cmap='plasma')
plt.colorbar(label='count in bin')
```

```python
import seaborn as sns
sns.kdeplot(x,y,shade=True,cmap='Blues')
```
